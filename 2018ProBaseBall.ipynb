```python
import json
from collections import OrderedDict
 
# Ready for data
play_data = OrderedDict()

play_data["ver"] = "v0.01"
play_data["site"] = "Naver"
play_data["play_info"] = "2018ProBaseball"
play_data["language"] = "kor"
play_data["play_list"] = []

def strongDeal(txt) :
    if "<em>" in txt :
        txt = str(game_list[j]).replace("<strong class=\"td_score\">", "").replace("</strong>", "").replace("<em>", "").replace("</em>", "")
        return txt
    else :
        txt = str(game_list[j]).replace("<strong class=\"td_score\">", "").replace("</strong>", "").replace("<em class=\"vs\">", "").replace("</em>", "")
        return txt

from bs4 import BeautifulSoup as BS
import urllib.request as req

for i in range(3, 10) :
    url = "http://sports.news.naver.com/kbaseball/schedule/index.nhn?date=20180301&month="+str(i).zfill(2)+"&year=2018&teamCode="
    res = req.urlopen(url)
    soup = BS(res, "html.parser")
    play_list = soup.select("#calendarWrap > div > table > tbody > tr")
    date = ""

    for i in play_list :
        game_list = []
        line = ""

        game_list.append(i.select_one("td > span.td_date > strong"))
        game_list.append(i.select_one("td > span.td_hour"))
        game_list.append(i.select_one("td > span.td_none"))
        game_list.append(i.select_one("td > span.team_lft"))
        game_list.append(i.select_one("td > strong.td_score"))
        game_list.append(i.select_one("td > span.team_rgt"))
        game_list.append(i.select_one("td > span.cancel"))

        play_score = OrderedDict()
        for j in range(len(game_list)) :
            if j==0 :
                if game_list[j] is not None :
                    date = game_list[j].string
                line = line + date
                play_score["Date"] = "2018"+date.split(".")[0].zfill(2)+date.split(".")[1].zfill(2)
                
            if j!=0 and ('NoneType' not in str(type(game_list[j]))) :
                if j==1 :
                    line = line + " " + game_list[j].string
                    play_score["Hour"] = game_list[j].string
                if j==2 :
                    line = line + " " + game_list[j].string
                    play_score["None"] = game_list[j].string
                if j==3 :
                    line = line + " " + game_list[j].string
                    play_score["Team_lft"] = game_list[j].string
                if j==4 :
                    line = line + " " + strongDeal(str(game_list[j]))
                    play_score["Score"] = strongDeal(str(game_list[j]))
                if j==5 :
                    line = line + " " + game_list[j].string
                    play_score["Team_rgt"] = game_list[j].string
                if j==6 :
                    line = line + " " + game_list[j].string
                    play_score["Cancel"] = game_list[j].string
                    
        play_data["play_list"].append(play_score)
        print(line)
```

```python
# Write JSON
with open('2018Baseball.json', 'w', encoding="utf-8") as make_file:
    json.dump(play_data, make_file, ensure_ascii=False, indent="\t")
```
